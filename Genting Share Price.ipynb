```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

df=pd.read_excel(r"C:\Users\weije\Desktop\3182.KL.xlsx")
print(df)
Open_genting=df["Open"]
Close_genting=df["Close"]
Date=df["Date"]

closing=df[df.Close=="4.43"]
print(closing)
ax1=plt.subplot(1,2,1)
plt.plot(Open_genting,Date,marker="o")
plt.xlabel("Opening Genting")
plt.ylabel("Date")
ax1.set_xticks(Open_genting)
ax1.set_xticklabels(Date)
plt.title("Relationship between Genting Share Price and Date")

ax2=plt.subplot(1,2,1)
plt.plot(Close_genting,Date,marker="o",color="red")
plt.xlabel("Closing Genting")
plt.ylabel("Date")
ax1.set_xticks(Close_genting)
ax1.set_xticklabels(Date)
plt.title("Relationship between Genting Share Price and Date")
plt.show()
```
